## Графіки декількох функцій

Побудуйте графіки $ f(x) = x^2 $ та $ g(x) = -x^2 + 5 $ на одній площині у діапазоні $ x = [-3, 3] $.



```python
# Ваш код тут:
x = var('x')
f(x) = x^2
g(x) = -x^2 + 5

p1 = plot(f(x), (x, -3, 3), color='blue', legend_label='f(x) = x²')
p2 = plot(g(x), (x, -3, 3), color='red', legend_label='g(x) = -x² + 5')

graph = p1 + p2
graph.show(title='Графіки функцій f(x) та g(x)', axes_labels=['x', 'y'])
```
