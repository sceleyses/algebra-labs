### 1. Абелеві групи




1. Напишіть функцію, яка для заданого $n$ повертає кількість попарно неізоморфних абелевих груп порядку $n$.
2. Знайдіть перші сто значень $n$, для яких існує рівно одна абелева група порядку $n$, і порівняйте з розкладом $n$ у добуток простих.
3. Напишіть гіпотезу про те, для яких $n$ існує рівно одна абелева група порядку $n$.
4. Спробуйте знайти всі $n$, для яких існує рівно $n$ абелевих груп порядку $n$.

```python
#№1
def count_abelian_groups(n):
    if n == 1:
        return 1
    # Отримуємо розклад на множники: список пар (просте, степінь)
    factors = list(factor(n))
    count = 1
    for p, exponent in factors:
        # Множимо на кількість розбиттів степеня (number_of_partitions)
        count *= number_of_partitions(exponent)
    return count

# Перевірка
print(f"Кількість абелевих груп порядку 4: {count_abelian_groups(4)}")  
print(f"Кількість абелевих груп порядку 6: {count_abelian_groups(6)}")  
print(f"Кількість абелевих груп порядку 8: {count_abelian_groups(8)}")   

#№2
found_count = 0
n = 1
results = []

print(f"\n{'n':<5} | {'Розклад на множники'}")
print("-" * 30)

while found_count < 100:
    if count_abelian_groups(n) == 1:
        results.append(n)
        # Виводимо перші 10 для наочності, щоб не засмічувати екран
        if found_count < 100: 
            print(f"{n:<5} | {factor(n)}")
        found_count += 1
    n += 1

#№3
print("\nГіпотеза:\nІснує рівно одна абелева група порядку n тоді і тільки тоді,коли у розкладі на прості множники всі показники ступеня дорівнюють 1")

#№4

solutions = []
for n in range(1, 10001):
    if count_abelian_groups(n) == n:
        solutions.append(n)

print(f"Знайдені розв'язки(перевірено до 10000): {solutions}")
```

### 3. Корінь з одиниці за модулем $p$

1. Знайдіть перші сто простих чисел $p$, для яких факторкільце $\mathbb{Z}_p[x]/(x^2+1)$ є полем.  $\qquad$ % PolynomialRing(GF(p))
2. Знайдіть остачі при ділення цих простих чисел на $m$ для деяких значень $m=2,3,\ldots,10$.        
3. Спробуйте знайти закономірність і сформулювати гіпотезу про такі прості числа.


```python
#№1
# Знаходимо перші 100 простих чисел p, для яких p mod 4 == 3
count = 0
p = 3  # Починаємо з першого простого числа, що підходить
prime_list = []

while count < 100:
    if is_prime(p) and p % 4 == 3:
        prime_list.append(p)
        count += 1
    p = next_prime(p)

print("Перші 100 простих чисел p, для яких Zp[x]/(x^2+1) є полем (p mod 4 == 3):")
print(prime_list)

#№2
for m in range(2, 11):
    remainders = set()
    for p in prime_list:
        remainders.add(p % m)

    # Упорядковуємо та виводимо
    sorted_remainders = sorted(list(remainders))
    print(f"\nОстачі простих чисел за модулем m = {m}:")
    print(f"p mod {m} належить: {sorted_remainders}")
    
#№3
print("\nГіпотеза:\nПросте число p(крім p = 2) задовільняє умову, що Z_p[x]/(x^2+1) є полем тод і тільки тоді, коли p ≡ 3 (mod 4)")
```

### 4. Ідеали та факторкільця кільця многочленів

1. Задайте кільце многочленів $R=\mathbb{Z}_{7}[x]$. $\qquad$  %  R.\<x\>=Integers(7)[]
3. Задайте ідеали $I_1=(2x^3+5x+1)$, $I_2=(x^4+3x^3+2)$, $I_3=(x^5+x^3+x+1)$.  $\qquad$    %  I = R.ideal(f(x))
4. Побудуйте факторкільця $R/I$ і знайдіть його порядок. $\qquad$  %  R.quotient(І)
5. Перевірте, для яких ідеалів факторкільце $R/I$ є полем.
6. Перевірте, чи має елемент $(x^5+3x^2+2)+I$ обернений у факторкільці $R/I$ та знайдіть обернений.

```python
#№1
R.<x> = Integers(7)[]
print(f"Кільце: {R}")

#№2
f1 = 2*x^3 + 5*x + 1
f2 = x^4 + 3*x^3 + 2
f3 = x^5 + x^3 + x + 1

I1 = R.ideal(f1)
I2 = R.ideal(f2)
I3 = R.ideal(f3)

print(f"Ідеал 1: {I1}")
print(f"Ідеал 2: {I2}")
print(f"Ідеал 3: {I3}")

#№3
Q1 = R.quotient(I1)
Q2 = R.quotient(I2)
Q3 = R.quotient(I3)

print(f"Порядок R/I1: {Q1.cardinality()}") 
print(f"Порядок R/I2: {Q2.cardinality()}") 
print(f"Порядок R/I3: {Q3.cardinality()}") # 7^5
```
